```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

result = []
for year in range(2004, 2016):
    data = pd.read_csv('PredictionRUT'+ str(year) + '.csv')
    mse = 0.0
    for index,row in data.iterrows():
        mse += (row['rutung1'] - row['rutung2'])*(row['rutung1'] - row['rutung2'])
    result += [mse/len(data)]
fig = plt.figure()

ax1 = fig.add_subplot(111)

ax1.set_title("Difference between ratings")    
ax1.set_xlabel('MSE')
ax1.set_ylabel('Year')

ax1.plot(result, range(2004,2016), c='r', label='MSE')

leg = ax1.legend()

fig.savefig('mse.png')
plt.show()
```

```python

data2010 = pd.read_csv('PredictionRUT2010.csv')
real_results = pd.read_csv('RegularSeasonCompactResults.csv')
real_results = real_results[real_results['Season'] == 2010]
data2010['wins'] = 0
for index, row in data2010.iterrows():
    wins = 0
    for ind, real in real_results.iterrows():
        if (real['Wteam'] == row['Team'] or real['Lteam'] == row['Team']):
            if (real['Wteam'] > real['Lteam'] and real['Wteam'] == row['Team']):
                wins += 1
            if (real['Wteam'] < real['Lteam'] and real['Lteam'] == row['Team']):  
                wins += 1
    row['wins'] = wins
```

```python
fig2 = plt.figure()

ax2 = fig2.add_subplot(111)

ax2.set_title("Difference between real wins and prediction rating in 2010")    
ax2.set_xlabel('Prediction')
ax2.set_ylabel('Real wins')

ax2.plot(data2010['rutung1'], data2010['wins'], 'ro')

leg = ax2.legend()

fig.savefig('eeet.png')
plt.show()
```
